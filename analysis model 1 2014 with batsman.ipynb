```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.cross_validation import KFold   #For K-fold cross validation
from sklearn.utils import shuffle
```

```python
auc2014 = pd.read_csv('2014_auctions.csv') 
role = pd.read_csv('player_role2014.csv') 
```

```python
balls = []
runs = []
avg = []
srate = []
batsman = []
base_price  = []
buy_price  = []
for index, row in auc2014.iterrows():
    
    
    
    
    
    pname = str(row['player_name'].replace(' ','+'))
    #print pname
    prole = role[role['pname'].str.find(pname) != -1] 
    if(len(prole) == 0):
        continue
    prole = prole.iloc[0]['role']
 
    fname = 'score_files/2014_t20_' + pname + '.csv'
    #print fname
    if(prole == 0):
        continue
        
    if (os.path.isfile(fname)):
        data = pd.read_csv(fname)
        
        data['balls']  =  data['balls'].astype('str') 
        data = data[data['balls'] != '\xc2\xa0']
        
        data['balls']  =  data['balls'].astype('float') 
        data['runs']  =  data['runs'].astype('float') 
        
        #data = data[data['runs'] != '\xc2\xa0']
        if(len(data) < 2):
            continue
        bprice  = str(row['base_price'])   
        #bprice  = str(row['base_price'].replace(',',''))
        #bprice  = bprice.replace('$','')
        bprice = float(bprice)
        
        b2price  = str(row['buy_price'].replace(',',''))
        b2price  = b2price.replace('$','')
        b2price = float(b2price)
    
        batsman.append(pname)
        base_price.append(bprice) 
        buy_price.append(b2price)

        balls.append(np.sum(data['balls']))
        runs.append(np.sum(data['runs']))
        avg.append(np.sum(data['runs'])/(len(data)))
        srate.append(np.mean([(r/b)*100 for r,b in zip(data['runs'], data['balls'])]))
        
        
        
    else:
        print 'file no exists'
        print fname
        #continue
```

```python
import math
s2014 = pd.DataFrame()
s2014['batsman'] = batsman
s2014['base_price'] = base_price
s2014['buy_price'] = buy_price
s2014['runs'] = runs
s2014['balls'] = balls
s2014['avg'] = avg
s2014['srate'] = srate
pruns  = np.array(s2014['runs'])
pavg  = np.array(s2014['avg'])
psrate  = np.array(s2014['srate'])
pbase_price  = np.array(s2014['base_price'])
 

s2014['runs2'] = np.array([i*i for i in pruns])
s2014['avg2'] = np.array([i*i for i in pavg])
s2014['srate2'] = np.array([i*i for i in psrate])
s2014['base_price2'] = np.array([i*i for i in pbase_price])
s2014['runs3'] = np.array([math.sqrt(i) for i in pruns])
s2014['runs4'] = np.array([i*i*i for i in pruns])
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
   

    ## Note: does not handle mix 1d representation
    #if _is_1d(y_true): 
    #    y_true, y_pred = _check_1d_array(y_true, y_pred)

    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```
