```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.cross_validation import KFold   #For K-fold cross validation
from sklearn.utils import shuffle
```

```python
auc2012 = pd.read_csv('2012_auctions.csv') 
role = pd.read_csv('player_role2012.csv') 
```

```python
auc2012[:5]
```

```python
balls = []
runs = []
avg = []
srate = []
batsman = []
base_price  = []
buy_price  = []
for index, row in auc2012.iterrows():
    
    
    
    
    
    pname = str(row['player_name'].replace(' ','+'))
    #print pname
    prole = role[role['pname'].str.find(pname) != -1] 
    if(len(prole) == 0):
        continue
    prole = prole.iloc[0]['role']
 
    fname = 'score_files/2012_t20_' + pname + '.csv'
    #print fname
    if(prole == 0):
        continue
        
    if (os.path.isfile(fname)):
        data = pd.read_csv(fname)
        
        data['balls']  =  data['balls'].astype('str') 
        data = data[data['balls'] != '\xc2\xa0']
        
        data['balls']  =  data['balls'].astype('float') 
        data['runs']  =  data['runs'].astype('float') 
        
        #data = data[data['runs'] != '\xc2\xa0']
        if(len(data) < 2):
            continue
            
        bprice  = str(row['base_price'].replace(',',''))
        bprice  = bprice.replace('$','')
        bprice = float(bprice)
        
        b2price  = str(row['buy_price'].replace(',',''))
        b2price  = b2price.replace('$','')
        b2price = float(b2price)
    
        batsman.append(pname)
        base_price.append(bprice) 
        buy_price.append(b2price)

        balls.append(np.sum(data['balls']))
        runs.append(np.sum(data['runs']))
        avg.append(np.sum(data['runs'])/(len(data)))
        srate.append(np.mean([(r/b)*100 for r,b in zip(data['runs'], data['balls'])]))
        
        
        
    else:
        print 'file no exists'
        print fname
        #continue
```

```python
import math
s2012 = pd.DataFrame()
s2012['batsman'] = batsman
s2012['base_price'] = base_price
s2012['buy_price'] = buy_price
s2012['runs'] = runs
s2012['balls'] = balls
s2012['avg'] = avg
s2012['srate'] = srate
pruns  = np.array(s2012['runs'])
pavg  = np.array(s2012['avg'])
psrate  = np.array(s2012['srate'])
pbase_price  = np.array(s2012['base_price'])
 

s2012['runs2'] = np.array([i*i for i in pruns])
s2012['avg2'] = np.array([i*i for i in pavg])
s2012['srate2'] = np.array([i*i for i in psrate])
s2012['base_price2'] = np.array([i*i for i in pbase_price])
s2012['runs3'] = np.array([math.sqrt(i) for i in pruns])
s2012['runs4'] = np.array([i*i*i for i in pruns])
```

```python
import sklearn
from sklearn import linear_model
s2012.dropna(inplace = True)
len_data = len(s2012)
train_set = s2012[:int(len_data*0.8)]
len(train_set)
test_set = s2012[len(train_set):]
print (len(test_set['buy_price']))



def mean_absolute_percentage_error(y_true, y_pred): 
   

    ## Note: does not handle mix 1d representation
    #if _is_1d(y_true): 
    #    y_true, y_pred = _check_1d_array(y_true, y_pred)

    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
from sklearn.tree import DecisionTreeRegressor

train_set = s2012[:int(len_data*0.8)]
len(train_set)
test_set = s2012[len(train_set):]


model = DecisionTreeRegressor(criterion='mse',max_depth=5, min_samples_split=2, min_samples_leaf= 2)
ans =  'buy_price'
cols = ['runs','avg2','base_price','srate']
a = train_set[cols]
model.fit(a,train_set[ans])
test_feat = test_set[cols]

vals = np.array(model.predict(test_feat))
preds = model.predict(test_set[cols])
print len(vals)
print mean_absolute_percentage_error(np.array(test_set[ans]), np.array(preds))
```
